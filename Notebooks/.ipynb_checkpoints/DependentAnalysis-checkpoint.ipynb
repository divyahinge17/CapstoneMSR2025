```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = './CSV/DependentsCounts.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
min_value = df['DependentArtifacts'].min()
max_value = df['DependentArtifacts'].max()
print(f"Min DependentArtifacts: {min_value}, Max DependentArtifacts: {max_value}")

# Define bins based on the data range
bins = list(range(0, max_value + 5000, 5000))  # Adjust bin size based on data range
labels = [f'{i}-{i + 5000}' for i in bins[:-1]]  # Label for each bin

# Categorize the DependentArtifacts into the bins
df['DependentArtifactsRange'] = pd.cut(df['DependentArtifacts'], bins=bins, labels=labels, right=False)

# Count how many artifacts fall into each bin
count_per_range = df['DependentArtifactsRange'].value_counts().sort_index()

# Remove empty bins (if you prefer not to show them)
count_per_range = count_per_range[count_per_range > 0]

# Plotting the bar chart with a logarithmic scale for better visualization
plt.figure(figsize=(12, 6))
count_per_range.plot(kind='bar', color='skyblue')
plt.title('Count of Artifacts by DependentArtifacts Range')
plt.xlabel('DependentArtifacts Range')
plt.ylabel('Count of Artifacts (Log Scale)')
plt.yscale('log')  # Apply log scale to the y-axis
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.grid(True)
plt.show()
```
