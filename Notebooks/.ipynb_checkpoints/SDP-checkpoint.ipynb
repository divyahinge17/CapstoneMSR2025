```python
import pandas as pd
import os
import ast
import json
```

```python
file_path = './CSV/dependenciesDetails'
stability_file = './CSV/stabilityScore.xlsx'
```

```python
df_stability = pd.read_excel(stability_file)


sdp_data = []
for i, file in enumerate(os.listdir(file_path)):
    
    if i >= 11:
        break
    df = pd.read_csv(os.path.join(file_path, file))
    
    artifact = df['releaseID'].apply(lambda x: x.replace('"', '').rsplit(':', 1)[0])[0]
    print(artifact)

    df['dependencyArtIDs'] = df['dependencyArtIDs'].apply(json.loads)
    unique_dependencies = set(dependency for dep_list in df['dependencyArtIDs'] for dependency in dep_list)
    print (unique_dependencies)

    if not unique_dependencies:
        continue
    
    artifact_row = df_stability[df_stability['Artifact'] == artifact]
    artifact_stability = artifact_row['stabilityScore'].values[0]
    artifact_dependencies_count = artifact_row['DependenciesCount'].values[0]

    dependency_stabilities = df_stability[df_stability['Artifact'].isin(unique_dependencies)]

    lower_stability_count = dependency_stabilities[dependency_stabilities['stabilityScore'] < artifact_stability].shape[0]

    # Calculate SDP
    sdp = lower_stability_count / artifact_dependencies_count if artifact_dependencies_count > 0 else 0

    # Collect final data
    sdp_data.append({
        'Artifact': artifact,
        'DependentReleases': artifact_row['DependentReleases'].values[0],
        'DependentArtifacts': artifact_row['DependentArtifacts'].values[0],
        'DependenciesCount': artifact_dependencies_count,
        'StabilityScore': artifact_stability,
        'SDP': sdp
    })

    


            
```

```python
print(sdp_data)
```

```python
df_final = pd.DataFrame(sdp_data)
```
