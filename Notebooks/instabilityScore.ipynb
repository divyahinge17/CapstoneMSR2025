```python
import pandas as pd
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
file_path1 = './CSV/DependentsCounts.csv'
file_path2 = './CSV/DependenciesCountperArtifact.csv'
df_dependents = pd.read_csv(file_path1)
df_dependencies = pd.read_csv(file_path2)
```

```python
df_dependents
```

```python
df_dependencies
```

```python
df_score = pd.merge(df_dependents,df_dependencies, on = 'Artifact', how = 'outer')
df_score.fillna(0, inplace=True)
```

```python
df_score
```

```python
df_score['InstabilityScore'] = df_score['DependenciesCount']/(df_score['DependenciesCount']+df_score['DependentArtifacts'])
```

```python
df_score
```

```python
#excel_file_path = './CSV/stabilityScore.xlsx'
#df_score.to_excel(excel_file_path, index=False)
```

```python
# df after removing artifacts that have stability score of 1
df_filtered = df_score[df_score['InstabilityScore'] != 1]
df_filtered = df_filtered.reset_index(drop=True)
```

```python
df_filtered
```

```python
sns.histplot(df_filtered['InstabilityScore'], kde=True,bins = 20 ,linewidth=0)
sns.kdeplot(df_filtered['InstabilityScore'], color='blue', linewidth=2)

# Customizing the plot
plt.title('Distribution of Instability Score')
plt.xlabel('Instability Score')
plt.ylabel('Density')
plt.show()

# Explanation of KDE curve:
# A KDE (Kernel Density Estimate) curve is a smooth, continuous line that represents the probability density function 
# of a dataset. It helps to visualize the distribution of data points and is useful for understanding the underlying 
# trends or patterns. In this plot, the KDE curve shows how the stability scores are distributed across different values.

# Calculating the average stability score
avg_stability = df_filtered['InstabilityScore'].mean()

# Creating a new column to categorize scores as 'Above Average' or 'Below Average'
df_filtered['Above_Avg'] = df_filtered['InstabilityScore'] > avg_stability

# Counting values above and below the average
counts = df_filtered['Above_Avg'].value_counts().rename(index={True: 'Above Avg', False: 'Below Avg'})

# Plotting the counts
sns.barplot(x=counts.index, y=counts.values, palette="viridis")

# Customizing the plot
plt.title('Count of Instability Scores Above and Below Average')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
numerical_columns = ["DependentReleases", "DependentArtifacts", "DependenciesCount", "InstabilityScore"]
correlation_matrix = df_filtered[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.set(style='white', font_scale=1.2)

g = sns.JointGrid(data= df_filtered, x='InstabilityScore', y='DependentReleases', xlim=(0, 1), ylim=(0, 300000), height=5)
g = g.plot_joint(sns.regplot, color="xkcd:muted blue")
g = g.plot_marginals(sns.distplot, kde=False, bins=12, color="xkcd:bluey grey")
plt.tight_layout()
```
